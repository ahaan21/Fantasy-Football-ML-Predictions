```python
# Generate 2022 Prediction Outputs

rb_df_22 = pd.read_excel("/content/gdrive/MyDrive/DVA Project/Data/Predictions/Raw_Predictions/2022_RB_predictions_2022-04-11.xlsx")
wr_df_22 = pd.read_excel("/content/gdrive/MyDrive/DVA Project/Data/Predictions/Raw_Predictions/2022_WR_predictions_2022-04-12.xlsx")
qb_df_22 = pd.read_excel("/content/gdrive/MyDrive/DVA Project/Data/Predictions/Raw_Predictions/2022_QB_predictions_2022-04-11.xlsx")
te_df_22 = pd.read_excel("/content/gdrive/MyDrive/DVA Project/Data/Predictions/Raw_Predictions/2022_TE_predictions_2022-04-12.xlsx")
```

```python
keep_cols = ['Player','Max_Team_Name','Pos','Current_Year','Rank_pred', 'y_hat']

qb_preds_21 = qb_df.query("Cooks == False and Boxcox == False and PCA == True and Model == 'EN' and Scoring == False and One_Year == True and cutoff_val == 150 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
qb_preds_21 = qb_preds_21[keep_cols]

qb_preds_22 = qb_df_22.query("Cooks == False and Boxcox == False and PCA == True and Model == 'EN' and Scoring == False and One_Year == True and cutoff_val == 150 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
qb_preds_22['Rank_pred'] = qb_preds_22.reset_index().index + 1
qb_preds_22 = qb_preds_22[keep_cols]

print(qb_preds_21.head())
print(qb_preds_22.head())
```

```python
keep_cols = ['Player','Max_Team_Name','Pos','Current_Year','Rank_pred', 'y_hat']

rb_preds_21 = rb_df.query("Cooks == True and Boxcox == False and PCA == True and Model == 'RF' and Scoring == False and One_Year == True and cutoff_val == 50 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
rb_preds_21 = rb_preds_21[keep_cols]

rb_preds_22 = rb_df_22.query("Cooks == True and Boxcox == False and PCA == True and Model == 'RF' and Scoring == False and One_Year == True and cutoff_val == 50 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
rb_preds_22['Rank_pred'] = rb_preds_22.reset_index().index + 1
rb_preds_22 = rb_preds_22[keep_cols]
rb_preds_21.head()
rb_preds_22.head()
```

```python
keep_cols = ['Player','Max_Team_Name','Pos','Current_Year','Rank_pred', 'y_hat']

wr_preds_21 = wr_df.query("Cooks == False and Boxcox == False and PCA == False and Model == 'RF' and Scoring == False and One_Year == False and cutoff_val == 150 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
wr_preds_21 = wr_preds_21[keep_cols]

wr_preds_22 = wr_df_22.query("Cooks == False and Boxcox == False and PCA == False and Model == 'RF' and Scoring == False and One_Year == False and cutoff_val == 150 and Metric == 'Sum'").sort_values('y_hat', ascending=False)
wr_preds_22['Rank_pred'] = wr_preds_22.reset_index().index + 1
wr_preds_22 = wr_preds_22[keep_cols]

print(wr_preds_21.head())
print(wr_preds_22.head())
```

```python
keep_cols = ['Player','Max_Team_Name','Pos','Current_Year','Rank_pred', 'y_hat']

te_preds_21 = te_df.query("Cooks == False and Boxcox == True and PCA == False and Model == 'EN' and Scoring == True and One_Year == False and cutoff_val == 9 and Metric == 'Avg'").sort_values('y_hat', ascending=False)
te_preds_21['y_hat'] = te_preds_21['y_hat'] * 17
te_preds_21 = te_preds_21[keep_cols]

te_preds_22 = te_df_22.query("Cooks == False and Boxcox == True and PCA == False and Model == 'EN' and Scoring == True and One_Year == False and cutoff_val == 9 and Metric == 'Avg'").sort_values('y_hat', ascending=False)
te_preds_22['y_hat'] = te_preds_22['y_hat'] * 17
te_preds_22['Rank_pred'] = te_preds_22.reset_index().index + 1
te_preds_22 = te_preds_22[keep_cols]

print(te_preds_21.head())
print(te_preds_22.head())
```

```python
# Join together all positions for 2022 predictions with Satwik's file
prod_data = pd.read_excel('/content/gdrive/MyDrive/DVA Project/Data/Pivoted_Data/pivoted_player_plus_team_data_ZEROS_Alex_Version_2022.xlsx')
```

```python
prod_rb = prod_data[prod_data['Pos'] == 'RB']
prod_wr = prod_data[prod_data['Pos'] == 'WR']
prod_te = prod_data[prod_data['Pos'] == 'TE']
prod_qb = prod_data[prod_data['Pos'] == 'QB']

prod_rb_merged = prod_rb.merge(rb_preds_22, left_on=['Player','Current_Year','Pos', 'Max_Team_Name'], right_on=['Player','Current_Year','Pos', 'Max_Team_Name'], how='left', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
prod_wr_merged = prod_wr.merge(wr_preds_22, left_on=['Player','Current_Year','Pos', 'Max_Team_Name'], right_on=['Player','Current_Year','Pos', 'Max_Team_Name'], how='left', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
prod_qb_merged = prod_qb.merge(qb_preds_22, left_on=['Player','Current_Year','Pos', 'Max_Team_Name'], right_on=['Player','Current_Year','Pos', 'Max_Team_Name'], how='left', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
prod_te_merged = prod_te.merge(te_preds_22, left_on=['Player','Current_Year','Pos', 'Max_Team_Name'], right_on=['Player','Current_Year','Pos', 'Max_Team_Name'], how='left', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
```

```python
predictions_22 = pd.concat([prod_rb_merged, prod_wr_merged, prod_te_merged, prod_qb_merged])
predictions_22 = predictions_22.sort_values(['Rank_pred', 'y_hat'], ascending=[True, False])

predictions_22.to_excel('/content/gdrive/MyDrive/DVA Project/Data/Predictions/2022_Predictions.xlsx')
```
